```python
import json
import csv
import sys
import pandas as pd
import matplotlib.pyplot as plt

data = []
with open('/Users/vaibhavigaekwad/Desktop/R Shiny Project/yelp-dataset/yelp_academic_dataset_review.json', encoding="utf8") as f:
    for line in f:
        data.append(json.loads(line))
        
len(data)
```

```python
reviewid = []
businessid= []
userid = []
stars = []
text = []
date = []


for entry in range(0, len(data)):     
    reviewid.append(data[entry]['review_id'])
    businessid.append(data[entry]['business_id'])
    userid.append(data[entry]['user_id'])
    stars.append(data[entry]['stars'])
    text.append(data[entry]['text'])
    date.append(data[entry]['date'])
    
    
data = {'review_id':reviewid,'business_id':businessid,'user_id':userid,'stars':stars,'text':text,'date':date}
review_data  = pd.DataFrame(data)
#review_data = open('review_data.csv','w') as csv_file:
#    csv_writer = csv.writer(review_data) 

#data.close()


#data = sys.argv[1]          #check if file empty or not
#review_data = sys.argv[2]
```
